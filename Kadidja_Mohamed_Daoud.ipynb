```python
import json
from google.colab import files

# Importer le fichier JSON depuis l'ordinateur
imported = files.upload()

# Lire le nom du fichier Json import√©
nom_fichier = next(iter(imported.keys()))

# Lire et afficher le contenu du fichier JSON
with open(nom_fichier, "r") as fichier_json:
    contenu_fichier = json.load(fichier_json)
    # Afficher le contenu avec une indentation
    print(json.dumps(contenu_fichier, indent=4))
```
